Задание 1

```python
import pandas as pd
import numpy as np

with open('horse_data.names', 'r') as reader:
  print(reader.read())
```

```python
df = pd.read_csv('horse_data.csv', header = None)
df.head()
```

```python
horse_data = pd.DataFrame({
    'Surgery': pd.to_numeric(df.iloc[:, 0], errors='coerce'),
    'Age': pd.to_numeric(df.iloc[:, 1], errors='coerce'),
    'Rectal Temperature': pd.to_numeric(df.iloc[:, 3], errors='coerce'),
    'Pulse': pd.to_numeric(df.iloc[:, 4], errors='coerce'),
    'Respiratory rate': pd.to_numeric(df.iloc[:, 5], errors='coerce'),
    'Temperature of extremities': pd.to_numeric(df.iloc[:, 6], errors='coerce'),
    'Pain': pd.to_numeric(df.iloc[:, 10], errors='coerce'),
    'Outcome': pd.to_numeric(df.iloc[:, 22], errors='coerce')
    })

horse_data['Age'].replace(9, 2, inplace=True)
horse_data.head()
```

Задание 2

```python
def statistic_show(column_name):
  df_1 = horse_data.dropna(subset=column_name)
  mode = df_1[column_name].mode()[0]
  q1 = df_1[column_name].quantile(0.25)
  q3 = df_1[column_name].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - (1.5 * iqr)
  upper_bound = q3 + (1.5 * iqr)
  outliers = df_1[~df_1[column_name].between(lower_bound, upper_bound, inclusive=True)]
  print('Mode = ', mode)
  print('LOWER = ', lower_bound)
  print('UPPER = ', upper_bound)
  print(df_1[column_name].describe())
  return outliers

def statistic_show_2(column_name):
  mode = horse_data[column_name].mode()[0]
  values = horse_data[column_name].value_counts()
  uq = horse_data[column_name].unique()
  print('Mode = ', mode)
  print('Unique: ', uq)
  return values
```

```python
#Rectal temperature
#Выбросы незначительны
statistic_show('Rectal Temperature')
```

```python
#Respiratory rate
#Выбросы мб значительны
statistic_show('Respiratory rate')
```

```python
#Pulse
#Выбросы также мб значительны
statistic_show('Pulse')
```

```python
#Age
statistic_show_2('Age')
```

```python
#Surgery
statistic_show_2('Surgery')
```

```python
#Temperature of extremities
statistic_show_2('Temperature of extremities')
```

```python
#Pain
statistic_show_2('Pain')
```

```python
#Outcome
statistic_show_2('Outcome')
```

```python
print(horse_data[(horse_data['Temperature of extremities'].isnull()) & (horse_data['Rectal Temperature'].isnull())])
#Оставим те сторки, в которых заполнена хотя бы половина информации
horse_data_clear = horse_data.copy()
horse_data_clear = horse_data_clear.dropna(thresh=5)
print(horse_data.describe())
print('-------------------')
print(horse_data_clear.describe())
```

```python
#Исходя из данных, у нас есть показатели, которые не являются непрерывными величинами - outcome, surgery.
#Также пропуск в данных столбцах невелик.
#Можно попробовать замену модой.
#Столбец Temperature of extremities, Pain категориальные, поэтому предпочтительным методом будет также заполнение пропусков модой
fill_mode = horse_data_clear.copy()
fill_mode['Surgery'] = fill_mode['Surgery'].fillna(horse_data['Surgery'].mode()[0])
fill_mode['Outcome'] = fill_mode['Outcome'].fillna(horse_data['Outcome'].mode()[0])
print(fill_mode['Surgery'].value_counts())
print('--------------------')
print(fill_mode['Outcome'].value_counts())
```

```python
#Исходя из описания, частота дыхания является непрерывной величиной, имеются выбросы, а также нет взаимосвязи с другими параметрами.
#Поэтому можно попробовать использовать замену медианой
fill_median = fill_mode.copy()
fill_median['Respiratory rate'] = fill_median['Respiratory rate'].fillna(horse_data['Respiratory rate'].median())
print(horse_data['Respiratory rate'].describe())
print('-------------------------')
print(fill_median['Respiratory rate'].describe())
```

```python
res = horse_data.groupby('Temperature of extremities').agg({'Rectal Temperature': ['min','max']})
res
```

```python
#Столбец Temperature of extremities, Pain категориальные, поэтому предпочтительным методом будет также заполнение пропусков модой
#При заполнении пустых значений в поле pain нужно проверить взаимосвязь между surgery и outcome.
#Например, эвтаназию станут делать более вероятно при высоком уров

fill_mode_2 = fill_median.copy()
fill_mode_2['Temperature of extremities'] = fill_median['Temperature of extremities'].fillna(horse_data['Temperature of extremities'].mode()[0])
fill_mode_2['Pain'] = fill_mode_2['Pain'].fillna(fill_mode_2.groupby(['Outcome', 'Surgery'])['Pain'].transform(lambda x: x.mode()[0]))
print(horse_data['Pain'].describe())
print('-------------------------')
print(fill_mode_2['Pain'].describe())
```

```python
#Можно предположить, что показатель rectal tempretarute и temperatures of extremities взаимосвязаны.
#Также исвестно, что температура повышается при инфекции, что может быть после операции
#Т.к выбросы у показателя rectal temperature имеются, то заполнение будет по медиане
print(horse_data.groupby(['Temperature of extremities','Surgery'])['Rectal Temperature'].median())

fill_median_by_rectemp = fill_mode_2.copy()
fill_median_by_rectemp['Rectal Temperature'] = fill_median_by_rectemp['Rectal Temperature'].fillna(fill_median_by_rectemp.groupby(['Temperature of extremities','Surgery'])['Rectal Temperature'].transform('median'))
print(horse_data['Rectal Temperature'].describe())
print('-------------------------')
print(fill_median_by_rectemp['Rectal Temperature'].describe())
```

```python
#Итоговым пустым значением остается показатель pulse.
#Из описания данных, можно предположить, что пульс имеют связб с возрастом, а также уровнем боли (при высокой боли, пульс тоже повышается)
#Pulse является также непрерывным значением
print(horse_data.groupby(['Age', 'Pain'])['Pulse'].median())

filled_data = fill_median_by_rectemp.copy()
filled_data['Pulse'] = filled_data['Pulse'].fillna(filled_data.groupby(['Age', 'Pain'])['Pulse'].transform('median'))
print(horse_data['Pulse'].describe())
print('-------------------------')
print(filled_data['Pulse'].describe())
```
